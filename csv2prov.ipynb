Converting a CSV file to PROV
=============================

This document provides a generic example of converting a csv file into a PROV document

Assumptions:

* a CSV file is considered a prov:collection of type csv_file
* the CSV file name is used as the prov:label
* a row is considered a prov:collection of type csv_row
* the header row contains the name of a prov:type exentension  
* the first column in the header contains unique IDs for each row  
* the index of each row/record is defined as an integer in prov:location  
* the value in each cell is a prov:entity of the given column header type  
* the IDs for all prov:collection and prov:entity are UUIDs

```python
import os
from socket import gethostname
from uuid import uuid1

from IPython.core import display
import pandas as pd
import prov.model as prov
```

```python
def csv2provgraph(filename, namespace=None, n_rows=500):
    """
    filename: path to file
    namespace: prov.Namespace instance to map column names to
    n_rows: number of rows to process
    """
    if namespace is None:
        namespace = prov.Namespace("ex","http://example.com#")
    
    def create_entity(graph, column, row_index, value):
        column = column.replace(" ", "_") # people still put spaces in column headers, really?
        obj_attr = [(prov.PROV["type"], namespace[column])] + \
                   [(prov.PROV["label"], "%s" % namespace[column]),
                    (prov.PROV["location"], "%s" % row_index),
                    (prov.PROV["value"], "%s" % value)
                   ]
        id = get_id()
        graph.entity(id, obj_attr)
        return id
    
    # create a new graph
    g = prov.ProvBundle()
    g.add_namespace(namespace)
    
    # uuid method
    get_id = lambda : namespace[uuid1().hex]
    
    # csv prov:collection
    csv_id = get_id()
    csv_collection = g.collection(csv_id)
    csv_collection.add_extra_attributes({prov.PROV['type']: namespace['csv_file'],
                                         prov.PROV['label']:filename}
                                       )
    
    data = pd.read_csv(filename)
    columns = data.keys()
    row_count = 0
    for row in data.iterrows():
        if row_count >= n_rows: 
            break
        row_count +=1
        row_id = get_id()
        # each row is a collection
        row_collection = g.collection(row_id)
        row_collection.add_extra_attributes({prov.PROV['type']: namespace['csv_row']})
        g.hadMember(csv_collection, row_id)       
        for column in columns:
            row_index = row[0]
            value = row[1][column]
            value_id = create_entity(g, column, row_index, value)
            g.hadMember(row_collection, value_id)
             
    return g
```

```python
nidm = prov.Namespace('nidm','http://nidm.nidash.org#')
g = csv2provgraph("NKI.1-39.phenotypic.csv", namespace=nidm, n_rows=500)
```

```python
with open('csvprov.provn', 'wt') as fp:
    fp.writelines(g.get_provn())
import json
json.dump(g, open('csvprov.json', 'wt'), cls= prov.ProvBundle.JSONEncoder)
```
